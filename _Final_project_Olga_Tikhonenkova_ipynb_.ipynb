```python
import pandas as pd
df = pd.read_csv("/content/Supermarket Sales 2.csv")
#df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Adding a column 'Revenue'
df["Revenue"] = df['Unit price'] * df['Quantity']
df.head(3)
```

```python
#Total revenue:
total_revenue = df['Revenue'].sum()
print('Total revenue = ', total_revenue)
```

```python
# Average receipt amount:
average_receipt = df['Revenue'].mean().round(2)
print('Average receipt: ', average_receipt)
```

```python
# Average check amount per day:

revenue_per_day = df.groupby('Date')['Revenue'].sum()
revenue_per_day
average_check_per_day = total_revenue / 365
print('Average check per day = :', average_check_per_day.round(2))
```

```python
#Top product categories:
df_top = (df.groupby('Product line', as_index=False)['Revenue'].sum().sort_values(by = 'Revenue', ascending=False))
df_top
```

```python
# The share of goods in revenue:
#type(df_top_10)
df_top['Share of goods'] = df_top['Revenue'] / total_revenue
df_top
```

```python
# Time-based sales analysis: days
# Convert the 'Date' column from string to datetime
try:
    df['Date'] = pd.to_datetime(df['Date'])
    print("Date column successfully converted to datetime")
except Exception as e:
    print("Error converting Date column:", e)

daily_revenue = df.groupby('Date', as_index=False)['Revenue'].sum()
daily_revenue
```

```python
# Revenue during the week:
daily_revenue['Day_type'] = daily_revenue['Date'].dt.weekday.apply(lambda x: 'Weekend' if x >=5 else 'Workday')
daily_revenue
```

```python
# Time-based sales analysis: months
df['Month'] = df['Date'].dt.to_period('M').dt.to_timestamp()
#df.head(3)
monthly_revenue = (df.groupby(['Month','Branch'], as_index=False)['Revenue'].sum())
monthly_revenue
```

```python
# Time-based prices analysis: months

monthly_price = (df.groupby(['Month', 'Branch'], as_index=False)['Unit price'].mean())
monthly_price
```

```python
# Revenue by branches:
branch_revenue = (df.groupby('Branch', as_index=False)['Revenue'].sum())
branch_revenue
```

```python
# The share of cash and card payments:
cash_and_card = (df.groupby('Payment', as_index=False)['Revenue'].sum())
cash_and_card
```

```python
# The largest receipt:
top_receipt = df['Revenue'].idxmax()
df.loc[top_receipt]

```

```python
# Sales analysis by product line and branch:
analysis_prod_branch = (df.groupby(['Product line', 'Branch'], as_index=False)['Revenue'].sum())
analysis_prod_branch
```

```python
# Revenue by customer type monthly:
customer_month_revenue = (df.groupby(['Month', 'Customer type'], as_index=False)['Revenue'].sum())
customer_month_revenue
```

```python
# Revenue by gender monthly:
gender_month_revenue = (df.groupby(['Month', 'Gender'], as_index=False)['Revenue'].sum())
gender_month_revenue
```

```python
# Prices product lines analysis:
# Food & Beverages
df_food = df[df['Product line'] == 'Food & Beverages']
df_food
daily_price_food = (df_food.groupby(['Date', 'Branch'], as_index=False)['Unit price'].mean())
daily_price_food
```

```python
# Home & lifestyle
df_home = df[df['Product line'] == 'Home & Lifestyle']
df_home
daily_price_home = (df_home.groupby(['Date', 'Branch'], as_index=False)['Unit price'].mean())
daily_price_home
```

```python
# Electronics
df_electronics = df[df['Product line'] == 'Electronics']
df_electronics
daily_price_electronix = (df_electronics.groupby(['Date', 'Branch'], as_index=False)['Unit price'].mean())
daily_price_electronix
```

```python
# Sports & Travel
df_sport = df[df['Product line'] == 'Sports & Travel']
df_sport
daily_price_sport = (df_sport.groupby(['Date', 'Branch'], as_index=False)['Unit price'].mean())
daily_price_sport
```

```python
# Fashion & Accs
df_fashion = df[df['Product line'] == 'Fashion & Accs']
df_fashion
daily_price_fashion = (df_fashion.groupby(['Date', 'Branch'], as_index=False)['Unit price'].mean())
daily_price_fashion
```
